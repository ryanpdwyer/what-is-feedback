```python
# Bode plot utility functions
def dB_sm(x):
    return 20*sm.log(sm.Abs(x), 10)

def dB(x):
    return 20*np.log10(np.abs(x))

def phase_degrees(x):
    return 180/np.pi*np.unwrap(np.angle(x))

def adjust_y_ticks(ax, delta):
    ylim = np.array(ax.get_ylim()) / delta
    ymin = ylim[0]//1  # Round towards - infinity
    ymax = -(-ylim[1]//1)  # Trick to round towards + infinity
    # Note: this rounds away from zero so we never make the axis limits smaller
    ax_new_lim = np.array([ymin, ymax]) * delta
    ax_new_ticks = np.arange(ax_new_lim[0], ax_new_lim[1]+1, delta)
    ax.set_ybound(*ax_new_lim)
    ax.set_yticks(ax_new_ticks)

def find_gain(mag, gain):
    if mag.max() > gain:
        return np.max(np.flatnonzero(mag > gain))
    else:
        return 0
    
def unity_gain(mag):
    return find_gain(mag, 0)
    
def phase_margin(mag, phase):
    return ((phase[unity_gain(mag)] % 360) - 180)  

def bode_f(func, xlim, N):
    f_min = xlim[0]
    f_max = xlim[1]
    f = np.logspace(np.log10(f_min), np.log10(f_max), N)
    
    transfer = func(f)
    mag = dB(transfer)
    phase = 180/np.pi*np.unwrap(np.angle(transfer))
    return f, mag, phase

def bode_f_kwargs(func, xlim, N, **kwargs):
    f_min = xlim[0]
    f_max = xlim[1]
    f = np.logspace(np.log10(f_min), np.log10(f_max), N)
    
    transfer = func(f, **kwargs)
    
    try:
        mag = dB(transfer)
        phase = 180/np.pi*np.unwrap(np.angle(transfer))
    except:
        print(transfer)
        raise
        
    return f, mag, phase

def get_bode_s(laplace, s, xlim, N=1000):
    f = sm.Symbol('f')
    laplace_numpy = sm.lambdify([f], laplace.subs(s, 2j*np.pi*f), modules='numpy')
    
    return bode_f(laplace_numpy, xlim, N)

def bode_quantities(loop_transfer, var, xlim, N):
    numerical_transfer = sm.lambdify(var, loop_transfer, modules='numpy')
    
    return bode_f(numerical_transfer, xlim, N)
    

def bode_plot(loop_transfer, var, xlim, N=1000, gain_point=0, mag_lim=None, phase_lim=None, figax=None):
    data = f_range, magnitude, phase = bode_quantities(loop_transfer, var, xlim, N)
    
    gain_index = find_gain(magnitude, gain_point)
    print(u"Freq at gain: {:.2g}".format(f_range[gain_index]))
    print(u"Phase at gain: {:.0f}°".format(phase[gain_index]))
    
    return bode(f_range, magnitude, phase, gain_point, xlim, mag_lim, phase_lim, figax)

def bode(freq, mag, phase, gain_point=0, xlim=None, mag_lim=None, phase_lim=None, figax=None):
    """Make a nice bode plot for the given frequency, magnitude, and phase data.
    
    If provided, xlim is the xlimits to set on the plot.
    If provided, mag_lim is a tuple with (mag_min, mag_max, mag_delta),
    and phase is also a tuple of the same type."""
    gain_index = find_gain(mag, gain_point)
    with mpl.rc_context({'figure.figsize': (8,6), 'lines.linewidth': 1.5, 'figure.dpi':300, 'savefig.dpi': 300, 'font.size': 16,}):
        if figax is None:
            fig, (ax1, ax2) = plt.subplots(nrows=2)
        else:
            fig, (ax1, ax2) = figax
        ax1.yaxis.grid(True, linestyle='-', color='.8')
        ax2.yaxis.grid(True, linestyle='-', color='.8')
        ax1.semilogx(freq, mag)

        if xlim is not None:
            ax1.set_xlim(*xlim)
        if mag_lim is not None:
            ax1.set_ylim(mag_lim[0], mag_lim[1])
            adjust_y_ticks(ax1, mag_lim[2])
        else:
            adjust_y_ticks(ax1, 20)
    
        ax1.axvline(x=freq[gain_index], color='k',  linestyle='--')

        ax2.semilogx(freq, phase, linewidth=1.5)
        
        if xlim is not None:
            ax2.set_xlim(*xlim)
        if phase_lim is not None:
            ax2.set_ylim(phase_lim[0], phase_lim[1])
            adjust_y_ticks(ax2, phase_lim[2])
        else:
            adjust_y_ticks(ax2, 30)

        ax2.axvline(x=freq[gain_index], color='k',  linestyle='--')
        ax1.set_ylabel('Magnitude [dB]')
        ax2.set_ylabel('Phase [degrees]')
        ax2.set_xlabel('Frequency')
        fig.tight_layout()
        fig.show()
        return fig, (ax1, ax2)
    
def bode_plot_s(loop_transfer, var, xlim, N=1000, gain_point=0, mag_lim=None, phase_lim=None, figax=None):
    """Create a bode plot for a sympy transfer function, expressed in terms of var s,
    plotted over a frequency range given in xlim.
    
    Other parameters described below.
    
    N is the number of points to calculated the transfer function at.
    gain_point determines where to draw a line on the plot
        this can be used to highlight the unity gain bandwidth, or the -3 dB point of a filter.
    
    mag_lim gives the magnitude axis limits as a tuple (low, high, tick_spacing)
    phase_lim gives the phase axis limits as a tuple (low, high, tick_spacing)
    figax allows the figure to be drawn on axes you provide, rather than a new set of axes.
        It should be a nested tuple, (fig, (ax1, ax2))
    """
    f = sm.symbols('f', real=True)
    return bode_plot(loop_transfer.subs(var, 2*sm.pi*sm.I*f), f, xlim, N, gain_point, mag_lim, phase_lim, figax)

def bode_plot_f(func, xlim, N, gain_point=0, mag_lim=None, phase_lim=None, **kwargs):
    freq, magnitude, phase =  bode_f_kwargs(func, xlim, N, **kwargs)
    gain_index = find_gain(magnitude, gain_point)
    print(u"Freq at gain: {:.2g}".format(freq[gain_index]))
    print(u"Phase at gain: {:.0f}°".format(phase[gain_index]))
    return bode(freq, magnitude, phase, gain_point=gain_point, mag_lim=mag_lim, phase_lim=phase_lim)

def bode_plot_interact(loop_transfer, var, xlim, N=1000, gain_point=0, mag_lim=None, phase_lim=None, subs={}, plot_kwargs={}):
    """A utility function for interacting with sympy transfer functions. Put variables to vary in subs, with the tuple you would
    pass to interact (min, max, delta), such as,
    
    {omega_0: (0, 1, 0.01)}"""
    f = sm.symbols('f', real=True)
    loop_transfer_f = loop_transfer.subs(var, 2j*sm.pi*f)
    
    variables = [f]
    variables.extend(subs.keys())
    func = sm.lambdify(variables, loop_transfer_f, dummify=False, modules='numpy')
    
    args = inspect.getargspec(func).args
    args.remove('f')
    new_kwargs = {key:val for key, val in zip(args, subs.values())}
    
    def _interact_f(**kwargs):
        print(kwargs)
        if 'gain_point' in plot_kwargs:
            return bode_plot_f(func, xlim, N, mag_lim=mag_lim, phase_lim=phase_lim, **kwargs)
        else:
            return bode_plot_f(func, xlim, N, gain_point=gain_point, mag_lim=mag_lim, phase_lim=phase_lim, **kwargs)
    
    new_kwargs.update(plot_kwargs)
    
    interact(_interact_f, **new_kwargs)

def bode_df(df, invert_phase=False, gain_point=0, xlim=None, mag_lim=None, phase_lim=None, figax=None):
    """Make a bode plot of the Bode plot data in a pandas dataframe df, collected by the Digilent Waveforms ap
    
    Invert phase optionally inverts the phase
    gain_point draws a line at the highest frequency with gain greater than gain_point
    xlim sets the frequency limits of the plot
    mag_lim is a tuple (min, max, delta) setting the magnitude limits and tick marks
    phase_lim is a tuple (min, max, delta) setting the phase limits and tick marks
    figax allows putting the plot on an existing figure and axes.
        It should be passed as a nested tuple, figax=(fig, (ax1, ax2))
    """
    phase_invert = 0 if invert_phase is False else 180
    return bode(df["Frequency (Hz)"].values, df['Channel 2 (dB)'].values,
                np.unwrap(df['Channel 2 (deg)'].values - phase_invert, 180),
                gain_point=gain_point, xlim=xlim, mag_lim=mag_lim,
                phase_lim=phase_lim, figax=figax)
    
```

# Tools for continuous / discrete transfer functions

## Continuous: Laplace transform

For a continuous function $h(t)$,

$$H(s) = \int_0^{\infty} h(t) e^{-s t} dt$$

## Discrete: Z transform

For a sequence $h[n]$,

$$H(z) = \sum_0^{\infty} h[n] z^{-n}$$

In the z transform, $z^{-1}$ represents delaying the sequence 1 element.

## Continuous transfer function: polynomial in $s$

In continuous time, we Laplace transform a response function $h(t)$ to a transfer function $H(s)$. For typical linear circuit elements, the transfer function will be of the form,

$$
\begin{align}
H(s) = \frac{b_0 + b_1 s + ... + b_N s^N}{a_0 + a_1 s + ... + a_N s^N}
\end{align}
$$

$H(s)$ represents the solution of a differential equation in the time domain.

### Implementation note

In `scipy.signal` and Matlab, functions dealing with continuous time filters default to returning arrays of coefficients $b_n$, $a_n$ in descending order,

$$ ([b_N, b_{N-1}, ..., b_0], \; [a_N, a_{N-1}, ..., a_0])$$

For example, the transfer function of a first order low pass filter with a cutoff frequency of 1 Hz is,

$$H(s) = \frac{1}{\frac{s}{2\pi}+1}$$

We can create such a filter in `scipy.signal` using `signal.butter`


```python
lp = Continuous(*signal.butter(1, 2*np.pi, btype='low', analog=True))
lp.b, lp.a
```

This is the same filter shown above, but `scipy.signal` normalizes it so that the largest polynomial power is 1.

$$H(s) = \frac{1}{\frac{s}{2\pi}+1} = \frac{2 \pi}{s + 2 \pi}$$

# Discrete transfer function: polynomial in negative powers of $z$
In discrete time, we use the z transform, and write a transfer function as,

$$
\begin{align}
H(z) = \frac{y(z)}{x(z)} = \frac{b_0 + b_1 z^{-1} + ... + b_N z^{-N}}{a_0 + a_1 z^{-1} + ... + a_N z^{-N}}
\end{align}
$$

While the Laplace transform represents the solution to a differential equation relating the input $x$ to the output $y$, the z transform represents the solution to a difference equation relating the input to the output. Specifically,

$$a_0 y[n] = \sum_{k=0}^{N} b_k x[n-k] - \sum_{k=1}^N a_k y[n-k]$$

The output is based on convolving the input with the coefficients of $b$ and convolving the previous output with $a$.

Based on this difference equation, we can break digital filters into two types.

## IIR Filters

If the denominator of the transfer function is not $1$, the filter is an infinite impulse response filter (IIR). These filters have more in common with analog filters we are familiar with. We will see IIR filters later.

## FIR Filters

If the denominator of the transfer function is $1$, the transfer function becomes

$$
\begin{align}
H(z) = \frac{y(z)}{x(z)} = b_0 + b_1 z^{-1} + ... + b_N z^{-N}
\end{align}
$$

which is called a finite impulse response (FIR) filter. The output of an FIR filter can be calculated with the discrete convolution, $y[n] = b[n]*x[n]$. The DSP lock-ins use FIR filters on the X and Y channels (set using the time constant TC, filter slope of 6, 12, 18 or 24 dB/octave).

# Continuous to discrete

Since a transfer function in negative powers of $z$ corresponds to a difference equation, if we can transform $H(s)$ to $H(z)$, we can easily compute the output using the difference equation.

For a sampling time $T_s$, the Laplace variable $s$ becomes,

$$ z = e^{s T_s} $$

This is reasonable, because $z^{-1}$ represents delaying a sequence by one. $z^{-1} = e^{-s T_s}$ which corresponds to delaying our continuous time system by one sampling time.

To get a polynomial in negative powers of $z$ out, we need to approximate the exponential. A reasonable approximation to use,

$$ z = \frac{T_s}{2}\frac{1-z}{1+z}$$

is called the [bilinear transform](http://en.wikipedia.org/wiki/Bilinear_transform) (specifically Tustin's transformation).

# Example: Simulating cantilever dynamics

The cantilever responds to a force $f(t)$ with a displacement $x(t)$.

$x(t) = \chi(t)*f(t)$

Our cantilever has a transfer function,

$$\chi(s) = \frac{1}{k_c}\frac{1}{1+\frac{s}{Q \omega_0} + \frac{s^2}{\omega_0^2}}$$

with $Q \gg 1$. In the time domain, we have (approximately, requires $Q \gg 1$),

$$ \chi(t) = \frac{2 \pi f_c}{k_c} e^{-\pi f_c t/Q} \sin(2 \pi f_c t)$$

```python
chi_s = 1/k_c/(1+s/omega_0/Q + s**2/omega_0**2)
typical_cantilever = {k_c:1, Q:10000, omega_0:50*2*np.pi}
figax_wide = bode_plot_s(chi_s.subs(typical_cantilever), s, (10, 1000), phase_lim=(-210, 30, 60))
```

```python
figax_zoomed = bode_plot_s(chi_s.subs(typical_cantilever), s, (49.9, 50.1), phase_lim=(-180, 0, 30))
```

## Naive approach

Calculate the time-domain response function at discrete times, or the frequency domain response function at discrete frequencies, and convolve (in the time-domain) or multiply (in the frequency domain) the input with the response function.

```python
def chi(t, f_c, k_c, Q):
    """Time domain cantilever response function, for Q >> 1."""
    pi, exp, sin = np.pi, np.exp, np.sin
    return (2*pi*f_c/k_c) * exp(-pi*f_c*t/Q) * sin(2*pi*f_c*t)
```

```python
dt = 3e-4
t0 = np.arange(0, 500, dt)
chi0 = chi(t0, 50, 1, 10000)*dt
```

```python
fig, (ax1, ax2) = plt.subplots(2)
fig.set_figheight(7)
fig.set_figwidth(6)
ax1.plot(t0, chi0)
ax2.plot(t0, chi0)
ax1.set_xlabel("Time [ms]")
ax2.set_xlabel("Time [ms]")
ax1.set_ylabel("Response [m/N]")
ax2.set_ylabel("Response [m/N]")
ax2.set_xlim(0, 0.25)
fig.tight_layout()
```

You can see it takes a very long time for the response function to decay; we are using 1.67 million data points to represent the response of a simple system with only 1 parameter that isn't a scale factor. In the frequency domain, we can do multiplication rather than convolution, but we still need to sample large amounts of data points to accurately represent the transfer function near resonance.

```python
t01 = np.arange(-0.1, 500, 3e-4)
x01 = np.where(t01>0, 1, 0)
```

```python
y01 = np.zeros_like(t01)
# We can only calculate this output quickly because the step response
# is the sum of the impulse response. If we wanted to simulate the response to a more complicated input, this would take
# a very long time
y01[-chi0.size:] = np.cumsum(chi0)
```

The below plot shows the cantilever step response to a 1 nN force applied at $t=0$.

```python
fig, (ax1, ax2) = plt.subplots(2)
fig.set_figheight(7)
fig.set_figwidth(6)
ax1.plot(t01, x01, t01, y01)
ax2.plot(t01, x01, t01, y01)
ax1.set_xlabel("Time [ms]")
ax2.set_xlabel("Time [ms]")
ax1.set_ylabel("Displacement [nm]")
ax2.set_ylabel("Displacement [nm]")
ax1.set_ylim(-0.1, 2.1)
ax2.set_ylim(-0.1, 2.1)
ax1.set_xlim(0, 500)
ax2.set_xlim(-0.05, 0.25)
fig.tight_layout()
```

# Direct discretization

On the other hand, we can directly switch to a discrete difference equation. To do so, we must take $\chi(s)$, and turn it into a function $\chi(z)$. 

For a sampling time $T_s$, the Laplace variable $s$ becomes,

$$ z = e^{-s T_s} $$

A reasonable approximation to use,

$$ z = \frac{T_s}{2}\frac{1-z}{1+z}$$

is called the bilinear transform (specifically Tustin's transformation).

```python
# Substitute the pade approximation for s, then substitute the exact expression for z(s) to see how the appoximation
# affects the continuous time transfer function
chi_tustin_freq = chi_s.subs(typical_cantilever).subs(pade).subs({z:sm.exp(s*T_s)})
```

```python
# Ts specified in ms, time steps are 300 ns, 100 ns, 30 ns.
for Ts in (3e-4, 1e-4, 3e-5):
    bode_plot_s(chi_tustin_freq.subs(T_s, Ts), s, (49.9, 50.1), figax=figax_zoomed)
    figax_wide = bode_plot_s(chi_tustin_freq.subs(T_s, Ts), s, (10, 1000), phase_lim=(-210, 30, 60), figax=figax_wide)
```

### Simulating the output

```python
b, a, dt = discrete(chi_s.subs(typical_cantilever), s, z_i, 1e-4)
b, a
```

These arrays are very small; just 3 terms each.

```python
t1 = np.arange(-0.01, 500, dt)
x1 = np.where(t1 > 0, 1, 0)
y1 = signal.lfilter(b, a, x1)
```

This is much faster to calculate. Let's look at the response to a force on resonance as well.

```python
def cantilever_response_f(f):
    t2 = t1
    # Resonance frequency slightly shifted by the time step
    x2 = np.sin(2*np.pi*f*t2)
    y2 = signal.lfilter(b, a, x2) / 1000
    fig2, ax21 = plt.subplots()
    fig2.set_figheight(6)
    fig2.set_figwidth(8)
    ax21.plot(t2, x2*5, t2, y2, linewidth=1.5)
    ax21.hlines(0, 0, 500)
    ax21.set_ylim(-10, 10)
    ax21.set_xlabel("Time [ms]")
    ax21.set_ylabel(u"Displacement [µm]")
    ax21.set_xlim(450, 450.1)
    fig2.tight_layout()
```

The cantilever displacement is in green, the input force (1 nN force, arbitrarily scaled) in blue.

# Cantilever Example: Conclusions

- In the naive approach, we truncated the time domain response of our cantilever.
    - We were approximating our cantilever response function with an FIR (finite impulse response) filter. We needed hundreds of thousands or millions of taps to get a good approximation to our cantilever, because the cantilever time domain response decays very slowly.
- If we directly convert our continuous time transfer function to a discrete time transfer function, we get an infinite impulse response (IIR) filter. We only need 3 coefficients each in the numerator and denominator to get a very good approximation to our cantilever, and can very quickly simulate the response of the system.
- This is an extreme example, but it holds true whenever you want to simulate the output of a continuous time system.

# Summary

- Discrete time systems important for simulation, and because cantilever signal / frequency is converted from analog to digital and back in our experiments
- Simulation
    - `scipy.signal` and Matlab have useful functions for dealing with transfer functions, simulating output
    - Worthwhile to transform $T(s) \rightarrow T(z)$ rather than just evaluating the continuous time transfer function at discrete values
    - For complicated IIR transfer functions, worry about stability! Matlab has features to implement IIR filters without affecting stability, as does the bleeding edge version of [scipy](https://github.com/scipy/scipy/pull/3717).
- Digital instruments used in our group
    - DSP lock-ins use *FIR* digital filters which are very different from analog filters.
        - This needs to be taken into account, especially when a lock-in is part of a feedback loop
    - I'm not sure what the PLLpro uses; it may use IIR filters (which act much more like analog filters)
    - FPGA: obviously this would involve digital signal processing

# Bank balances, calculated using an IIR filter

Deposits $x[n]$, interest rate $r$, balance $y[n]$.

\begin{align}
y[n]& = x[n] + (1+r) y[n-1]& \\

\\

We could analyze these transfer functions for stability as well (the account balance is stable if interest rate $r < 0$).
